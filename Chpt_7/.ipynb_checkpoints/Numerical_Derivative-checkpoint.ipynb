# The Numerical Derivative

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 2*x**2
```

```python
x = np.arange(0, 5, 0.001)
y = f(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
p2_delta = 0.0001
x1 = 2
x2 = x1 + p2_delta
```

```python
y1 = f(x1)
y2 = f(x2)

print((x1, y1), (x2, y2))
```

```python
approx_derivative = (y2-y1)/(x2-x1)
b = y2 - approx_derivative * x2
```

```python
def tangent_line(x):
    return approx_derivative * x + b
```

```python
to_plot = [x1-0.9, x1, x1+0.9]
plt.plot(to_plot, [tangent_line(i) for i in to_plot])

print('Approx derivative for f(x)', f'where x = {x1} is {approx_derivative}')

plt.show()
```

```python
plt.figure()
plt.plot(x, f(x))
plt.scatter(x1, y1, color='C1', s=50)
plt.plot(to_plot, [tangent_line(i) for i in to_plot], 'C1--', linewidth = 2)
```
