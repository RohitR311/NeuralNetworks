# Tangents at various points

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 2 * x ** 2
```

```python
x = np.array(np.arange(0, 5, 0.001))
y = f(x)

colors = ["k", "g", "r", "b", "c"]
```

```python
def approx_tangent_line(x, approx_derivative):
    return (approx_derivative * x) + b
```

```python
for i in range(5):
    p2_delta = 0.0001
    x1 = i
    x2 = x1 + p2_delta

    y1 = f(x1)
    y2 = f(x2)

    print((x1, y1), (x2, y2))
    approx_derivative = (y2 - y1) / (x2 - x1)
    b = y2 - (approx_derivative * x2)

    to_plot = [x1 - 0.9, x1, x1 + 0.9]

#     plt.scatter(x1, y1, c=colors[i])
#     plt.plot(
#         [point for point in to_plot],
#         [approx_tangent_line(point, approx_derivative) for point in to_plot],
#         c=colors[i],
#     )

    plt.figure()
    plt.plot(x, f(x))
    plt.scatter(x1, y1, c=colors[i], s=50)
    plt.plot(
        [point for point in to_plot],
        [approx_tangent_line(point, approx_derivative) for point in to_plot],
        c=colors[i],
    )

    print("Approx derivative for f(x)", f"where x = {x1} is {approx_derivative}")

    plt.show()
```
