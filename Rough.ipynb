```python
from Code_Sets.Layers import Layer_Dense
```

```python
from Code_Sets.Activation_Functions import Activation_ReLU, Activation_Linear
```

```python
from Code_Sets.Losses import MeanSquaredError_Loss
```

```python
from Code_Sets.Optimizers import Optimizer_Adam
```

```python
from Code_Sets.Accuracies import Accuracy_Regression
```

```python
from Code_Sets.Model_Object import Model
```

```python
X, y = sine_data()
X_test, y_test = sine_data(100)
```

```python
X.shape
```

```python
model = Model()
```

```python
model.add(Layer_Dense(1, 64))
model.add(Activation_ReLU())
model.add(Layer_Dense(64, 64))
model.add(Activation_ReLU())
model.add(Layer_Dense(64, 1))
model.add(Activation_Linear())
```

```python
model.set(
    loss=MeanSquaredError_Loss(),
    optimizer=Optimizer_Adam(learning_rate=0.005, decay=1e-3),
    accuracy=Accuracy_Regression(),
)
```

```python
model.finalize()
```

```python
model.train(X, y, validation_data=(X_test, y_test), epochs=10000, print_every=100)
```
